# MATPLOTLIB

```python
import pandas as pd
import numpy as np
from matplotlib import pylab as pl
from matplotlib import pyplot as plt
import matplotlib.cm as cm

print(pd.__version__)
print(np.__version__)
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
a = np.linspace(0,5,10)
x = a
y = a*a
```

```python
read = df["reading score"]
write = df["writing score"]
```

```python
pl.plot(x,y,"r")
```

```python
pl.subplot(1,2,1)
pl.plot(x,y,"r--")
pl.subplot(1,2,2)
pl.plot(y,x,"g*--")
```

```python
fig = plt.figure()
axis = fig.add_axes([0.5,0.5,1,1])
axis.plot(x,y,"g")
```

```python
fig,axis = plt.subplots(nrows=1,ncols=3)
for a in axis:
  a.plot(x,y,"g")
```

```python
fig1 = plt.figure(figsize=(16,9), dpi=100)
fig1.add_subplot()
plt.plot(x,y,"r*--")
```

```python
fig,axis = plt.subplots(figsize = (16,9), dpi=100)
axis.plot(x,y,"g*--")
axis.set_xlabel("Shravan X")
axis.set_ylabel("Shravan Y")
axis.set_title("Shravan graph")
```

```python
axis.legend(["Label1","Label2"])
fig,axis = plt.subplots(dpi=200)
axis.set_xlabel("Shravan X")
axis.set_ylabel("Shravan Y")
axis.set_title("Shravan graph")
axis.plot(x,x**5)
axis.plot(y,y**2)
axis.legend(["x","y"],loc=10)
```

```python
fig,axis = plt.subplots(dpi=200)
axis.plot(y,y+1,color="magenta", alpha = .3, linewidth=5)
axis.plot(y,y+2, color="cyan", linewidth=3)
axis.plot(y,y+3, linewidth=1)
```

```python
fig,axis = plt.subplots(dpi=200)
axis.plot(y,y+1,color="magenta", alpha = .3, lw=5, linestyle=":")
axis.plot(y,y+2, color="cyan", alpha = .6, lw=3, linestyle="-.")
axis.plot(y,y+3, color="black",lw=1, alpha = 1, linestyle="--")
```

```python
fig,axis = plt.subplots(dpi=200)
axis.plot(y,y+1,color="magenta", lw=3, marker="+")
axis.plot(y,y+2, color="cyan", lw=2, marker="o", markerfacecolor="red")
axis.plot(y,y+3, color="black",lw=1, marker="1", markersize=5)
```

```python
fig, Axis = plt.subplots(1,2,figsize=(16,9),dpi=100)
Axis[0].plot(x,x**2,x,x**3)
Axis[0].grid(True)
Axis[0].set_ylim([1,80])
Axis[1].plot(x,x**2,x,x**3)
Axis[1].set_xlim([0,3])
```

```python
n = np.array(range(1,21))
```

```python
fig,axis = plt.subplots(1,4, figsize=(16,4), dpi = 100)
axis[0].scatter(x,x**2)
axis[0].set_title("Scatter")

axis[1].step(x,x**2)
axis[1].set_title("Step")

axis[2].bar(x,x**2, align="center", width = 0.5)
axis[2].set_title("bar")

axis[3].fill_between(x,x**2)
axis[3].set_title("Fill Between")
```

```python
n = np.random.randn(100)
n1 = np.random.randn(1000)
fig, axis = plt.subplots(1,2, figsize=(16,8),dpi = 100)
axis[0].set_title("Histogram")
axis[0].hist(n, width = 0.4)
axis[1].set_title("Cumulative Histogram")
axis[1].hist(n1, width = 0.2, cumulative = True, bins = 100)
```
